```python
def fifo_page_replacement(pages, frame_size):
    frames = []
    page_faults = 0

    print("FIFO Page Replacement - pra3.ipynb:5")
    for page in pages:
        if page not in frames:
            page_faults += 1
            if len(frames) < frame_size:
                frames.append(page)
            else:
                frames.pop(0)   # Remove the oldest page
                frames.append(page)
        print(f"Page: {page} > Frames: {frames} - pra3.ipynb:14")

    print("Total Page Faults: - pra3.ipynb:16", page_faults)
```

```python
def lru_page_replacement(pages, frame_size):
    frames = []
    recent_use = []
    page_faults = 0

    print("\nLRU Page Replacement - pra3.ipynb:6")
    for page in pages:
        if page not in frames:
            page_faults += 1
            if len(frames) < frame_size:
                frames.append(page)
            else:
                lru_page = recent_use.pop(0)
                frames.remove(lru_page)
                frames.append(page)
        else:
            recent_use.remove(page)

        recent_use.append(page)
        print(f"Page: {page} > Frames: {frames} - pra3.ipynb:20")

    print("Total Page Faults: - pra3.ipynb:22", page_faults)
    # Fixed page reference string and frame size
pages = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3, 0, 3, 2]
frame_size = 3

fifo_page_replacement(pages, frame_size) # type: ignore
lru_page_replacement(pages, frame_size)
```
