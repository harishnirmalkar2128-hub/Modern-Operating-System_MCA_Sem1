```python
def bankers_algorithm(processes, available, max_need, allocation):
    n = len(processes)          # Number of processes
    m = len(available)          # Number of resources

    # Calculate Need matrix
    need = [[max_need[i][j] - allocation[i][j] for j in range(m)] for i in range(n)]

    finish = [False] * n
    safe_sequence = []
    work = available.copy()

    while len(safe_sequence) < n:
        found = False

        for i in range(n):
            if not finish[i]:
                # Check if needs can be satisfied
                if all(need[i][j] <= work[j] for j in range(m)):
                    # Allocate resources
                    for j in range(m):
                        work[j] += allocation[i][j]

                    safe_sequence.append(processes[i])
                    finish[i] = True
                    found = True

        if not found:
            break

    if len(safe_sequence) == n:
        print("System is in a SAFE state. - pra3.ipynb:31")
        print("Safe Sequence: - pra3.ipynb:32", " -> ".join(safe_sequence))
    else:
        print("System is in an UNSAFE state (Deadlock may occur). - pra3.ipynb:34")


# Example Input
processes = ["P0", "P1", "P2", "P3", "P4"]

available = [3, 3, 2]

max_need = [
    [7, 5, 3],
    [3, 2, 2],
    [9, 0, 2],
    [2, 2, 2],
    [4, 3, 3]
]

allocation = [
    [0, 1, 0],
    [2, 0, 0],
    [3, 0, 2],
    [2, 1, 1],
    [0, 0, 2]
]

bankers_algorithm(processes, available, max_need, allocation)
5
```
