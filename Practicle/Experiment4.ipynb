```python
from os import read
from urllib import request


def sstf(requests, head):
    total_seek = 0
    current = head
    pending = requests.copy()

    print("\nSSTF Disk Scheduling - pra3.ipynb:6")
    print("Head Movement: - pra3.ipynb:7")

    while pending:
        nearest = min(pending, key=lambda x: abs(x - current))
        movement = abs(nearest - current)
        total_seek += movement

        print(f"{current} > {nearest} (Seek = {movement}) - pra3.ipynb:14")

        current = nearest
        pending.remove(nearest)

    print("Total Seek Time: - pra3.ipynb:19", total_seek)
    return total_seek


# Example
sstf(request, read)
```

```python
def fcfs(requests, head):
    total_seek = 0
    current = head

    print("FCFS Disk Scheduling - pra3.ipynb:5")
    print("Head Movement: - pra3.ipynb:6")

    for req in requests:
        movement = abs(req - current)
        total_seek += movement
        print(f"{current} > {req} (Seek = {movement}) - pra3.ipynb:11")
        current = req

    print("Total Seek Time: - pra3.ipynb:14", total_seek)
    return total_seek


# Example
requests = [82, 170, 43, 140, 24, 16, 190]
head = 50

fcfs(requests, head)
```
