```python
import platform
import socket
import os

def system_info():
    print("=== Basic System Information ===\n - pra2.ipynb:6")

    # Operating system info
    print("Operating System: - pra2.ipynb:9", platform.system())
    print("OS Version: - pra2.ipynb:10", platform.version())
    print("OS Release: - pra2.ipynb:11", platform.release())

    # Machine and processor info
    print("Machine Type: - pra2.ipynb:14", platform.machine())
    print("Processor: - pra2.ipynb:15", platform.processor())

    # Python info
    print("Python Version: - pra2.ipynb:18", platform.python_version())

    # Network info
    print("Hostname: - pra2.ipynb:21", socket.gethostname())
    print("IP Address: - pra2.ipynb:22", socket.gethostbyname(socket.gethostname()))

    # User info
    print("Current User: - pra2.ipynb:25", os.getlogin())

if __name__ == "__main__":
    system_info()
```
