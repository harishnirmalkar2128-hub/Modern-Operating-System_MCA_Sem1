```python
from collections import deque

def round_robin(processes, quantum):
    time = 0
    queue = deque()
    remaining_bt = {p[0]: p[2] for p in processes}
    arrived = set()
    result = {}

    while remaining_bt:
        for pid, at, bt in processes:
            if at <= time and pid not in arrived:
                queue.append(pid)
                arrived.add(pid)

        if not queue:
            time += 1
            continue

        pid = queue.popleft()
        exec_time = min(quantum, remaining_bt[pid])
        remaining_bt[pid] -= exec_time
        time += exec_time

        if remaining_bt[pid] == 0:
            at = next(p[1] for p in processes if p[0] == pid)
            bt = next(p[2] for p in processes if p[0] == pid)
            ct = time
            tat = ct - at
            wt = tat - bt
            result[pid] = (ct, tat, wt)
            del remaining_bt[pid]
        else:
            queue.append(pid)

    return result


print("\nRound Robin Scheduling: - pra3.ipynb:39")
rr_result = round_robin(processes, quantum=2)
for pid, values in rr_result.items():
    print(f"P{pid}: CT={values[0]}, TAT={values[1]}, WT={values[2]} - pra3.ipynb:42")
```

```python
def priority_scheduling(processes):
    time = 0
    completed = []
    processes = processes.copy()

    while processes:
        available = [p for p in processes if p[1] <= time]
        if not available:
            time += 1
            continue

        pid, at, bt, pr = min(available, key=lambda x: x[3])
        time += bt
        ct = time
        tat = ct - at
        wt = tat - bt
        completed.append((pid, ct, tat, wt))
        processes.remove((pid, at, bt, pr))

    return completed


# Example with priority
priority_processes = [
    (1, 0, 5, 2),
    (2, 1, 3, 1),
    (3, 2, 8, 3)
]

print("\nPriority Scheduling: - pra3.ipynb:30")
for p in priority_scheduling(priority_processes):
    print(p)
```

```python
from platform import processor


def sjf(processes):
    time = 0
    completed = []
    processes = processes.copy()

    while processes:
        available = [p for p in processes if p[1] <= time]
        if not available:
            time += 1
            continue

        pid, at, bt = min(available, key=lambda x: x[2])
        time += bt
        ct = time
        tat = ct - at
        wt = tat - bt
        completed.append((pid, ct, tat, wt))
        processes.remove((pid, at, bt))

    return completed


print("\nSJF Scheduling: - pra3.ipynb:23")
for p in sjf(processor):
    print(p)
```

```python
def fcfs(processes):
    time = 0
    result = []

    for pid, at, bt in processes:
        if time < at:
            time = at
        ct = time + bt
        tat = ct - at
        wt = tat - bt
        time = ct
        result.append((pid, ct, tat, wt))

    return result


# Example
processes = [
    (1, 0, 5),
    (2, 1, 3),
    (3, 2, 8)
]

print("FCFS Scheduling: - pra3.ipynb:24")
for p in fcfs(processes):
    print(p)
```
