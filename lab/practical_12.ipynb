```python
def optimal_page_replacement(pages, frames):
    frame = []
    page_faults = 0

    print("Page\tFrames")

    for i in range(len(pages)):
        page = pages[i]

        # Page Hit
        if page in frame:
            print(page, "\t", frame)
            continue

        # Page Fault
        page_faults += 1

        # If frame has space
        if len(frame) < frames:
            frame.append(page)
        else:
            future = []
            for f in frame:
                if f in pages[i+1:]:
                    future.append(pages[i+1:].index(f))
                else:
                    future.append(float('inf'))

            # Replace page used farthest in future
            frame[future.index(max(future))] = page

        print(page, "\t", frame)

    print("\nTotal Page Faults:", page_faults)


# Example Input
pages = [7, 0, 1, 2, 0, 3, 0, 4]
frames = 3

optimal_page_replacement(pages, frames)
```
