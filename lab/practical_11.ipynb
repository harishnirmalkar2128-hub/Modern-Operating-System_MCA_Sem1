```python
# Memory Allocation using First Fit, Best Fit, Worst Fit

def first_fit(blocks, processes):
    print("\nFirst Fit Allocation:")
    for p in processes:
        allocated = False
        for i in range(len(blocks)):
            if blocks[i] >= p:
                print(f"Process {p} allocated to Block {blocks[i]}")
                blocks[i] -= p
                allocated = True
                break
        if not allocated:
            print(f"Process {p} not allocated")

def best_fit(blocks, processes):
    print("\nBest Fit Allocation:")
    for p in processes:
        best_index = -1
        for i in range(len(blocks)):
            if blocks[i] >= p:
                if best_index == -1 or blocks[i] < blocks[best_index]:
                    best_index = i
        if best_index != -1:
            print(f"Process {p} allocated to Block {blocks[best_index]}")
            blocks[best_index] -= p
        else:
            print(f"Process {p} not allocated")

def worst_fit(blocks, processes):
    print("\nWorst Fit Allocation:")
    for p in processes:
        worst_index = -1
        for i in range(len(blocks)):
            if blocks[i] >= p:
                if worst_index == -1 or blocks[i] > blocks[worst_index]:
                    worst_index = i
        if worst_index != -1:
            print(f"Process {p} allocated to Block {blocks[worst_index]}")
            blocks[worst_index] -= p
        else:
            print(f"Process {p} not allocated")

# Input
memory_blocks = [100, 500, 200, 300, 600]
processes = [212, 417, 112, 426]

# Calling functions
first_fit(memory_blocks.copy(), processes)
best_fit(memory_blocks.copy(), processes)
worst_fit(memory_blocks.copy(), processes)
```
