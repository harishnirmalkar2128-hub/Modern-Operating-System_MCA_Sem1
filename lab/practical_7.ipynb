```python
# Round Robin Scheduling Algorithm Implementation

processes = []
burst_time = []
remaining_time = []

n = int(input("Enter number of processes: "))

for i in range(n):
    processes.append(i + 1)
    bt = int(input(f"Enter Burst Time for Process P{i+1}: "))
    burst_time.append(bt)
    remaining_time.append(bt)

time_quantum = int(input("Enter Time Quantum: "))

time = 0
waiting_time = [0] * n
turnaround_time = [0] * n

# Keep track of last finish time to calculate waiting time
last_finish = [0] * n

# Round Robin Execution
while True:
    done = True

    for i in range(n):

        if remaining_time[i] > 0:
            done = False

            # If process needs more than time quantum
            if remaining_time[i] > time_quantum:
                time += time_quantum
                remaining_time[i] -= time_quantum
            else:
                # Process finishes
                time += remaining_time[i]
                waiting_time[i] = time - burst_time[i]
                remaining_time[i] = 0

    if done:
        break

# Turnaround Time = Waiting Time + Burst Time
for i in range(n):
    turnaround_time[i] = burst_time[i] + waiting_time[i]

# Output
print("\nProcess\tBurst Time\tWaiting Time\tTurnaround Time")
for i in range(n):
    print(f"P{processes[i]}\t{burst_time[i]}\t\t{waiting_time[i]}\t\t{turnaround_time[i]}")

avg_wt = sum(waiting_time) / n
avg_tat = sum(turnaround_time) / n

print(f"\nAverage Waiting Time: {avg_wt:.2f}")
print(f"Average Turnaround Time:Â {avg_tat:.2f}")
```
