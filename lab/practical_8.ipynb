```python
# Priority Scheduling (Non-Preemptive)

n = int(input("Enter number of processes: "))

processes = []
burst_time = []
priority = []

for i in range(n):
    processes.append(i + 1)
    bt = int(input(f"Enter Burst Time for Process P{i+1}: "))
    pr = int(input(f"Enter Priority for Process P{i+1} (Lower number = Higher priority): "))
    burst_time.append(bt)
    priority.append(pr)

# Combine process info for easy sorting
proc_info = list(zip(processes, burst_time, priority))

# Sort by priority (ascending â†’ higher priority first)
proc_info.sort(key=lambda x: x[2])

waiting_time = [0] * n
turnaround_time = [0] * n

# Calculate Waiting Time
for i in range(1, n):
    waiting_time[i] = waiting_time[i - 1] + proc_info[i - 1][1]

# Calculate Turnaround Time
for i in range(n):
    turnaround_time[i] = waiting_time[i] + proc_info[i][1]

# Display Result
print("\nProcess\tPriority\tBurst Time\tWaiting Time\tTurnaround Time")
for i in range(n):
    print(f"P{proc_info[i][0]}\t{proc_info[i][2]}\t\t{proc_info[i][1]}\t\t{waiting_time[i]}\t\t{turnaround_time[i]}")

# Averages
avg_wt = sum(waiting_time) / n
avg_tat = sum(turnaround_time) / n

print(f"\nAverage Waiting Time: {avg_wt:.2f}")
print(f"Average Turnaround Time: {avg_tat:.2f}")
```
