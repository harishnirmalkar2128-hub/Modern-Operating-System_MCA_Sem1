```python
def fifo_page_replacement(pages, frames):
    memory = []
    faults = 0

    for page in pages:
        if page not in memory:
            if len(memory) < frames:
                memory.append(page)
            else:
                memory.pop(0)      # Remove oldest page
                memory.append(page)
            faults += 1

        print("Page:", page, "Memory:", memory)

    print("Total Page Faults:", faults)


# Example
pages = [1, 3, 0, 3, 5, 6, 3]
frames = 3
fifo_page_replacement(pages, frames)

```
