```python
# Banker's Algorithm in Python (Simple Version)

# Number of processes and resources
n = 5   # Processes P0 to P4
m = 3   # Resources A, B, C

# Available resources
available = [3, 3, 2]

# Maximum resource requirement of each process
max_need = [
    [7, 5, 3],
    [3, 2, 2],
    [9, 0, 2],
    [2, 2, 2],
    [4, 3, 3]
]

# Allocated resources
allocation = [
    [0, 1, 0],
    [2, 0, 0],
    [3, 0, 2],
    [2, 1, 1],
    [0, 0, 2]
]

# Calculate Need matrix
need = [[max_need[i][j] - allocation[i][j] for j in range(m)] for i in range(n)]

finish = [False] * n
safe_sequence = []

work = available.copy()

while len(safe_sequence) < n:
    allocated = False
    for i in range(n):
        if not finish[i] and all(need[i][j] <= work[j] for j in range(m)):
            for j in range(m):
                work[j] += allocation[i][j]
            finish[i] = True
            safe_sequence.append(i)
            allocated = True
    if not allocated:
        break

# Output
if len(safe_sequence) == n:
    print("System is in SAFE state")
    print("Safe Sequence:", end=" ")
    for i in safe_sequence:
        print(f"P{i}", end=" ")
else:
    print("System is NOT in safe state")
```
