```python
# FCFS Disk Scheduling Algorithm

requests = [82, 170, 43, 140, 24, 16, 190]
head = 50

total_movement = 0
current = head

print("Head Movement:")
print(current, end="")

for req in requests:
    movement = abs(req - current)
    total_movement += movement
    current = req
    print(" ->", current, end="")

print("\nTotal Head Movement:", total_movement)
```
