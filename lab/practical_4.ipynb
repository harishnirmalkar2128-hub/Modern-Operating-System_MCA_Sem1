```python
import psutil

print("Top 5 Running Processes (Process Name + PID)")
print("-" * 45)

count = 0

for process in psutil.process_iter(['pid', 'name']):
    try:
        print(f"Process Name: {process.info['name']} \t PID: {process.info['pid']}")
        count += 1
        if count == 5:
            break
    except (psutil.NoSuchProcess, psutil.AccessDenied):
        pass
```
