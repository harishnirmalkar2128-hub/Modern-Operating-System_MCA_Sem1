```python
def predictive_lru(pages, capacity):
    cache = []
    freq = {}   # frequency of page usage

    print("Page\tCache\t\tStatus")

    for page in pages:
        # Update frequency
        freq[page] = freq.get(page, 0) + 1

        if page in cache:
            cache.remove(page)
            cache.append(page)
            print(f"{page}\t{cache}\tHIT")
        else:
            if len(cache) == capacity:
                # Predict next page (most frequent so far)
                predicted = max(freq, key=freq.get)

                # Evict LRU page unless it is predicted
                for p in cache:
                    if p != predicted:
                        cache.remove(p)
                        break

            cache.append(page)
            print(f"{page}\t{cache}\tMISS")

# Example
pages = [1, 2, 3, 2, 4, 1, 5, 2]
capacity = 3

predictive_lru(pages, capacity)
```
